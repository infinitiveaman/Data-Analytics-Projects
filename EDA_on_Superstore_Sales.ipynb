```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
import warnings
warnings.filterwarnings('ignore')

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,5)

print(" Libraries imported successfully!")

```

```python
from google.colab import files
import pandas as pd

print(" Please upload your dataset CSV file (e.g., 'SampleSuperstore.csv')")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
try:
    df = pd.read_csv(file_name, encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv(file_name, encoding='ISO-8859-1')
    except:
        df = pd.read_csv(file_name, encoding='cp1252')

print(f"\n Dataset '{file_name}' uploaded and loaded successfully!")
print("Dataset Shape:", df.shape)

df.head()
```

```python

print("Basic Info:")
df.info()

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

print("\n Cleaned dataset shape:", df.shape)
```

```python
# STATISTICAL SUMMARY

df.describe(include='all').T
```

```python

print("Number of unique customers:", df['Customer Name'].nunique() if 'Customer Name' in df.columns else "N/A")
print("Number of unique products:", df['Product Name'].nunique() if 'Product Name' in df.columns else "N/A")

if 'Region' in df.columns:
    print("Regions:", df['Region'].unique())
if 'Category' in df.columns:
    print("Categories:", df['Category'].unique())
```

```python

if 'Category' in df.columns:
    cat_sales = df.groupby('Category')[['Sales','Profit']].sum().reset_index()

    fig, ax = plt.subplots(1,2, figsize=(14,5))
    sns.barplot(x='Category', y='Sales', data=cat_sales, ax=ax[0], palette='Blues_d')
    ax[0].set_title("Total Sales by Category")

    sns.barplot(x='Category', y='Profit', data=cat_sales, ax=ax[1], palette='Greens_d')
    ax[1].set_title("Total Profit by Category")
    plt.show()

# Profit by Region
if 'Region' in df.columns:
    region_profit = df.groupby('Region')['Profit'].sum().reset_index()
    sns.barplot(x='Region', y='Profit', data=region_profit, palette='Purples_d')
    plt.title("Total Profit by Region")
    plt.show()

# Discount vs Profit
if 'Discount' in df.columns:
    sns.scatterplot(x='Discount', y='Profit', data=df, alpha=0.6, color='coral')
    plt.title("Discount vs Profit Relationship")
    plt.show()

#  Correlation Heatmap
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python

df['Sales'] = (
    df['Sales']
    .astype(str)
    .str.replace('[^\d\.]', '', regex=True)
    .astype(float)
)

if 'Profit' in df.columns:
    df['Profit'] = (
        df['Profit']
        .astype(str)
        .str.replace('[^\d\.-]', '', regex=True)
        .astype(float)
    )


if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
    df['Month'] = df['Order Date'].dt.to_period('M').astype(str)


df.dtypes.head()
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o')
plt.title(" Monthly Sales Trend")
plt.xticks(rotation=45)
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python

if 'Customer Name' in df.columns:
    top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
    sns.barplot(x=top_customers.values, y=top_customers.index, palette='mako')
    plt.title("Top 10 Customers by Total Sales")
    plt.show()

if 'Product Name' in df.columns:
    top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
    sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
    plt.title("Top 10 Products by Total Sales")
    plt.show()
```

```python

if 'Segment' in df.columns:
    seg_profit = df.groupby('Segment')[['Sales','Profit']].sum().reset_index()
    sns.barplot(x='Segment', y='Profit', data=seg_profit, palette='crest')
    plt.title("Profit by Segment")
    plt.show()

if 'Region' in df.columns and 'Category' in df.columns:
    pivot = df.pivot_table(values='Profit', index='Region', columns='Category', aggfunc='sum')
    sns.heatmap(pivot, annot=True, cmap='YlGnBu', fmt=".0f")
    plt.title("Profit Distribution by Region and Category")
    plt.show()
```

```python


# Profit margin
if 'Profit' in df.columns and 'Sales' in df.columns:
    profit_margin = (df['Profit'].sum()/df['Sales'].sum())*100
    print(f" Overall Profit Margin: {profit_margin:.2f}%")


```

```python

import matplotlib.ticker as mtick

#  Key metrics
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_discount = df['Discount'].mean() * 100 if 'Discount' in df.columns else 0
profit_margin = (total_profit / total_sales) * 100

print(" BUSINESS PERFORMANCE SUMMARY ")
print(f" Total Sales: ${total_sales:,.2f}")
print(f" Total Profit: ${total_profit:,.2f}")
print(f" Average Discount: {avg_discount:.2f}%")
print(f" Overall Profit Margin: {profit_margin:.2f}%")

# - Pie chart: Sales by Region
if 'Region' in df.columns:
    region_sales = df.groupby('Region')['Sales'].sum()
    plt.figure(figsize=(6,6))
    plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
    plt.title("Regional Sales Contribution")
    plt.show()

#  Pie chart: Sales by Category
if 'Category' in df.columns:
    category_sales = df.groupby('Category')['Sales'].sum()
    plt.figure(figsize=(6,6))
    plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("cool"))
    plt.title("Category-wise Sales Distribution")
    plt.show()

# Top 5 Sub-Categories by Profit
if 'Sub-Category' in df.columns:
    subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(5)
    sns.barplot(x=subcat_profit.values, y=subcat_profit.index, palette='flare')
    plt.title("Top 5 Sub-Categories by Profit")
    plt.xlabel("Profit ($)")
    plt.show()

# Profit Margin by Segment
if 'Segment' in df.columns:
    seg_margin = (df.groupby('Segment')[['Sales','Profit']].sum())
    seg_margin['Profit Margin (%)'] = (seg_margin['Profit'] / seg_margin['Sales']) * 100
    seg_margin = seg_margin.sort_values('Profit Margin (%)', ascending=False)

    sns.barplot(x=seg_margin['Profit Margin (%)'], y=seg_margin.index, palette='crest')
    plt.title("Profit Margin by Segment (%)")
    plt.xlabel("Profit Margin (%)")
    plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter())
    plt.show()
```
